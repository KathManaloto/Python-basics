# CONVERTING DATA TYPES

#### Converting String to Integer

```python
# int
num_int = 7
type(num_int)
```

```python
# stg
num_stg = '7'
type(num_stg)
```

```python
# Cannot add num_int and num_stg. Converting string to integer:
num_stg_conv = int(num_stg)
type(num_stg_conv)
```

```python
# Add num_int and the converted stg
num_sum = num_int + num_stg_conv
print(num_sum)
```

#### Converting List to Tuple

```python
# List
list_type = [1,2,3]
type(list_type)
```

```python
print(list_type)
```

```python
# Converting to tuple
tuple_type = tuple(list_type)
type(tuple_type)
```

```python
print(tuple_type)
```

#### Converting List to Set

```python
# Converting a list to a set removes duplicate values and keep only distinct elements.
list_type2 = [1,2,3,2,3,4,1,1,2,3,4,1]
type(list_type2)
```

```python
# Converting to set
set_type = set(list_type2)
type(set_type)
```

```python
print(set_type)
```

#### Converting Dictionary to a List

```python
# Example
dict_type = {'Name':'Hanabi', 'Age':20, 'Role':'Marksman'}
type(dict_type)
```

```python
dict_type.items() 
```

```python
dict_type.keys()
```

```python
dict_type.values()
```

```python
# Converting dictionary to list
dict_to_list = list(dict_type.values())
print(dict_to_list)
```

```python
dict_to_list = list(dict_type.keys())
print(dict_to_list)
```

```python
type(dict_to_list)
```
