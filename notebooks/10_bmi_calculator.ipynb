# BMI CALCULATOR

```python
# Formula: BMI = (weight in pounds * 703) / (height in inches * height in inches)
```

```python
# Input function

while True:
    name = input("Enter your name: ").strip()
    if name != "":
        break
    print("Please enter your name.")

while True:
    try:
        weight = float(input("Enter your weight in pounds: "))
        if weight > 0:
            break
        print("Invalid input! Weight must be greater than 0.")
        
    except ValueError:
        print("Please enter a valid input")

while True:
    try:
        height = float(input("Enter your height in inches: "))
        if height > 0:
            break
        print("Invalid input! Height must be greater than 0.")
        
    except ValueError:
        print("Please enter a valid input")
        
bmi = round(((weight * 703) / (height ** 2)),1)
print("BMI: " + str(bmi))

if (bmi < 18.5):
    print("Feedback: " + name + ", you are underweight. Maybe eat more nutritious foods and stay healthy!")
elif (bmi <= 24.9):
    print("Feedback: " + name + ", your BMI is normal. Great job, keep it up!")
elif (bmi <= 29.9):
    print("Feedback: " + name + ", you are overweight. Consider regular exercise and a balanced diet.")
elif (bmi <= 34.9):
    print("Feedback: " + name + ", you are in the obese range. It's important to monitor your weight and consult a healthcare professional.")
else:
    print("Feedback: " + name + ", you are morbidly obese. Please consult a healthcare professional as soon as possible.")
```
