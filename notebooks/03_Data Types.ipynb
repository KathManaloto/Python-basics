#### 4.) SETS

```python
# No duplicate values
daily_pints_log = {1,2,3,40,1,2,51,3}
print(daily_pints_log)
```

```python
print(daily_pints_log)
```

```python
# Compare 2 sets and determine unique values
daily_pints_log2 = {1,2,3,4,5,10,15,7,1,4,2,3,50}
```

```python
# Union
print(daily_pints_log | daily_pints_log2)
```

```python
# Intersection
print(daily_pints_log & daily_pints_log2)
```

```python
# Set difference
print(daily_pints_log - daily_pints_log2)
```

```python
# Symmetric difference
print(daily_pints_log ^ daily_pints_log2)
```

#### 5.) DICTIONARIES

```python
# Examples
hanabi_bio = {'Name':'Hanabi', 'Alias':'Scarlet Flower', 'Role':'Marksman'}
print(hanabi_bio)
```

```python
type(hanabi_bio)
```

```python
# Show dict keys
hanabi_bio.keys()
```

```python
# Show dict values
hanabi_bio.values()
```

```python
# Show all items in the dict
hanabi_bio.items()
```

```python
# Find value
hanabi_bio['Name']
```

```python
# Change value
hanabi_bio['Name'] = 'huāhuǒ'
```

```python
print(hanabi_bio)
```

```python
# Update values. 
#This will add the Age:20, update items that needs to be updated and keep other items with no changes
hanabi_bio.update({'Name': 'Hanabi', 'Age':20})
print(hanabi_bio)
```

```python
# Delete an item
del hanabi_bio['Age']
print(hanabi_bio)
```
