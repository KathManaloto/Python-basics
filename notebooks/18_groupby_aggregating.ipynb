# GROUP BY AND AGGREGATING

```python
# IMPORT PANDAS LIBRARY
import pandas as pd
```

```python
# READ IN FILE
df = pd.read_csv(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\pandas\Flavors.csv")
df
```

```python
# Performing group by in the 'Base Flavor' column
group_by_frame = df.groupby('Base Flavor')
```

```python
# Getting the mean 
group_by_frame.mean(numeric_only=True)
```

```python
# Alternatively
df.groupby('Base Flavor').mean(numeric_only = True)
```

```python
# Getting the count
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').size()
```

```python
# Getting the Min value
df.groupby('Base Flavor').min()
```

```python
# Getting the Max value
df.groupby('Base Flavor').max()
```

```python
# Getting the sum
df.groupby('Base Flavor').sum(numeric_only= True)
```

```python
# Getting multiple agg functions at once
df.groupby('Base Flavor').agg({'Flavor Rating':['mean', 'max','count','sum'], 'Texture Rating':['mean','max', 'count','sum']})
```

```python
# Grouping by multiple columns
df.groupby(['Base Flavor','Liked']).mean(numeric_only = True)
```

```python
# Grouping by multiple columns and getting multiple agg functions at once
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count','sum'],'Texture Rating':['mean','max','count','sum']})
```
