# FILTERING AND ORDERING

```python
# IMPORT PANDAS LIBRARY
import pandas as pd
```

```python
# READ IN DATA SET
df = pd.read_csv(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\pandas\world_population.csv")
df
```

#### FILTERING

```python
# Filter top 10 countries
df[df['Rank'] <= 10]
```

```python
# Filter by country
specific_countries = ["Bangladesh","Brazil"]
df[df['Country'].isin(specific_countries)]
```

```python
# Filter specific keyword
df[df['Country'].str.contains('United')]
```

```python
# Change the index to Country to make it more meaningful
df2 = df.set_index('Country')
df2
```

```python
# Filter specific columns
df2.filter(items = ['Continent', 'CCA2'], axis = 1)
```

```python
# Filter specific index
df2.filter(items = ['Zimbabwe'], axis = 0)
```

```python
# Filter rows with specific keyword
df2.filter(like = 'United', axis = 0)
```

```python
# Select row based on specific index label
df2.loc['United States']
```

```python
# Selects 4th row (row starts with 0)
df2.iloc[3]
```

```python
# Sort/Order by rank
df[df['Rank'] <= 10].sort_values(by='Rank', ascending = True)
```

```python
# Sort/Order by multiple columns
df[df['Rank'] <= 10].sort_values(by=['Continent','Country'], ascending = [False,True])
```
