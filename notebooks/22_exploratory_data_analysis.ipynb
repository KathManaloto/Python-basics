```python
# IMPORT LIBRARIES
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# READ IN DATASET
df = pd.read_csv(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\pandas\eda\world_population.csv")
df
```

```python
# FORMAT DECIMAL PLACES
pd.set_option('display.float_format',lambda x: '%.2f' % x)
df
```

```python
# LOOK INFO
df.info()
```

```python
# DESCRIBE
df.describe()
```

```python
# CHECK NO. OF NULL VALUES
df.isnull().sum()
```

```python
# CHECK UNIQUE VALUES
df.nunique()
```

```python
# TOP RANGE OF COUNTRIES (2022 Population)
df.sort_values(by = "2022 Population", ascending = False).head(10)
```

```python
# TOP 10 HIGHEST WORLD POPULATION PERCENTAGE
df.sort_values(by = "World Population Percentage", ascending= False).head(10)
```

```python
# CHECK CORRELATION BETWEEN COLUMNS
df.corr(numeric_only=True)
```

#### VISUALIZE DATA

```python
plt.rcParams['figure.figsize']=(20,7)
sns.heatmap(df.corr(numeric_only = True), annot = True)
plt.show()
```

#### GROUP DATA

```python
# Group by 'Continent'
df.groupby('Continent').mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
```

```python
# Show specific continent
df[df['Continent'].str.contains('Oceania')]
```

```python
df.columns
```

```python
# Visualize the grouped data
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
df2
```

```python
# This will result a graph, with continents as the rows and population as the data, which is incorrect
df2.plot()
```

```python
# Transposing df2 to make continents as columns
df3 = df2.transpose()
df3.plot()
```
