# DATA CLEANING IN PANDAS

```python
# IMPORT PANDAS LIBRARY
import pandas as pd
```

```python
# READ IN DATA SET
df = pd.read_excel(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\pandas\Customer Call List.xlsx")
df
```

#### CHECK FOR DATA THAT NEEDS TO BE CLEANED

```python
# REMOVE DUPLICATES 
df = df.drop_duplicates()
df
```

```python
# REMOVE UNNECESSARY COLUMNS
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df
```

#### CLEAN Last_Name COLUMN

```python
# Removing typos
df['Last_Name'] = df['Last_Name'].str.strip()
df['Last_Name'] = df['Last_Name'].str.lstrip('...')
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
```

```python
df
```

```python
# Alternatively, 
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
df
```

#### CLEAN Phone_Number COLUMN

```python
# Remove non-numeric values
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
df
```

```python
# Format the phone numbers. Need to change the data type of each row from phone_number column to string first
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))
df
```

```python
# Do the formatting using lambda
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# Remove nan-- and Na--
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--|Na--', '', regex = True)
df
```

#### CLEAN Address COLUMN

```python
# SPLIT THE ADDRESS COLUMN INTO 3 (Street_Address, State, Zip_Code)
df[["Street_Address", "State", "Zip_Code"]] = df['Address'].str.split(pat = ',',n = 2, expand = True)
df
```

#### CLEAN Paying Customer COLUMN

```python
# STANDARDIZE YES AND NO
df['Paying Customer'] = df['Paying Customer'].replace({'Y':'Yes','N':'No'})
df
```

#### CLEAN Do_Not_Contact COLUMN

```python
# STANDARDIZE YES AND NO
df['Do_Not_Contact'] =  df['Do_Not_Contact'].replace({'Y':'Yes','N':'No'})
df
```

#### REMOVE NaN and N/A FROM THE DATA FRAME

```python
# REMOVE NaN
df = df.fillna('')
df
```

```python
# REMOVE N/A
df = df.replace('N/a','')
df
```

#### FILTER DATA (phone numbers that can be contacted)

```python
# Filter Do_Not_Contact Column, remove Yes rows
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Yes':
        df.drop(x, inplace = True)

df
```

```python
# We can also use boolean filtering (df_filtered)
df_filtered = df[df['Do_Not_Contact'] != 'Yes']
df_filtered
```

```python
df
```

```python
# Filter Phone_Number Column, remove blank rows (df_filtered)
df_filtered= df_filtered[df_filtered['Phone_Number'] != '']
df_filtered
```

```python
# Filter Phone_Number Column, remove blank rows (df)
for x in df.index:
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x,inplace = True)
        
df

#OR df = df.dropna(subset = 'Phone_Number', inplace = True)
```
