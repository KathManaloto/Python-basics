# PANDAS VISUALIZATIONS

```python
# IMPORT MODULES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# READ IN DATASET
df = pd.read_csv(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\pandas\Ice Cream Ratings.csv")
df
```

```python
# SET Date AS INDEX
df = df.set_index('Date')
df
```

#### VISUALIZATION

```python
# Show all available stypes for visualization
print(plt.style.available)
```

```python
# Change style
plt.style.use('classic')
```

```python
# DEFAULT
df.plot()
```

```python
# LINE PLOT
df.plot(kind = 'line')
```

```python
# Break the plot by column
df.plot(kind = 'line', subplots = True)
```

```python
# Set Labels (x and y axes)
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
# BAR PLOT
df.plot(kind = 'bar')
```

```python
# Stacked Bar Plot
df.plot(kind = 'bar', stacked = True)
```

```python
# Plot specific column
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
# Horizontal bar chart
df.plot.barh(stacked = True)
```

```python
# Scatter Plot
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Yellow')
```

```python
# Histogram (default)
df.plot.hist()
```

```python
df.plot.hist(bins = 20)
```

```python
df.plot.hist(bins = 10)
```

```python
# Box Plot
df.boxplot()
```

```python
# Area Plot
df.plot.area()
```

```python
# Increase size
df.plot.area(figsize = (10,2))
```

```python
# Pie Chart
df.plot.pie(y = 'Flavor Rating', figsize = (10,7))
```
