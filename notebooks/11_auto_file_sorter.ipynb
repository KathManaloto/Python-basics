# AUTOMATIC FILE SORTER IN FILE EXPLORER

```python
# IMPORT MODULES
import os
import shutil
```

```python
# SET VARIABLES FOR SOURCE AND DESTINATION PATHS
source = os.path.join(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\file_sorter", "source")
destination = os.path.join(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\file_sorter", "destination")
```

```python
# DEFINE FILE CATEGORIES
categories = {
    "Images":[".png",".jpg",".jpeg",".gif"],
    "Documents":[".pdf",".txt",".docx",".xlsx",".csv"],
    "Audio":[".mp3",".wav"],
    "Video":[".mp4",".mkv"]
}
```

```python
# CREATE DESTINATION FOLDERS FOR EACH CATEGORY
for folder in list(categories.keys()) + ["Others"]:
    destination_path = os.path.join(destination, folder)

    if not os.path.exists(destination_path):
        os.makedirs(destination_path)
        print("Folder created: " + folder)
    else:
        print("Folder already created: " + folder)
```

```python
# List files from source folder
file_name = os.listdir(source)
file_name
```

```python
# MOVE FILES FROM SOURCE TO DESTINATION FOLDERS
for file in file_name:
    source_file_path = os.path.join(source, file)

    # Skip subdirectory
    if not os.path.isfile(source_file_path):
        continue

    # Split file name and extension
    name, ext = os.path.splitext(file)
    ext = ext.lower()
    
    # Initialized moved as False
    moved = False

    # Check each files and move to respective folders
    for folder, extensions in categories.items():
        if ext in extensions:
            shutil.move(source_file_path, os.path.join(destination, folder, file))
            print(f"Moved {file} -> {folder}")
            moved = True
            break

    if not moved:
        shutil.move(source_file_path, os.path.join(destination, "Others", file))
        print(f"Moved {file} → Others")
```
