# MERGE, JOIN AND CONCATENATE

```python
# IMPORT PANDAS LIBRARY
import pandas as pd
```

```python
# Read in first dataset (LOTR)
df1 = pd.read_csv(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\pandas\LOTR.csv")
df1
```

```python
# Read in second dataset (LOTR2)
df2 = pd.read_csv(r"C:\Users\kathm\OneDrive\Desktop\DataAnalystBootCamp\Python\pandas\LOTR 2.csv")
df2
```

#### MERGE

```python
# By default, it will perform inner join and merge on col with same name for both datasets
df1.merge(df2)
```

```python
# INNER JOIN (same as above example)
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
# OUTER JOIN
df1.merge(df2, how = 'outer')
```

```python
# LEFT JOIN
df1.merge(df2, how = 'left')
```

```python
# RIGHT JOIN
df1.merge(df2, how = 'right')
```

```python
# CROSS JOIN
df1.merge(df2, how = 'cross')
```

#### JOIN

```python
# Set the index for both datasets then join them
df4 = df1.set_index(['FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left', rsuffix = '_Right', how = 'outer')
df4
```

#### CONCATENATE

```python
# Concatenate df1 and df2
pd.concat([df1,df2])
```

```python
# Concatenate + inner join
pd.concat([df1,df2], join = 'inner')
```

```python
# Concatenate + outer join
pd.concat([df1,df2], join ='outer')
```

```python
# Concatenate + axis = 1
pd.concat([df1, df2], join = 'outer', axis = 1)
```
