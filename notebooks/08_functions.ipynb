#### ARGUMENTS

```python
# 1 argument
def num_squared(num):
    print(num ** 2)

# Calling the function
num_squared(5)
```

```python
# Multiple arguments
def num_squared_cust(num, power):
    print(num ** power)


num_squared_cust(5,3)
```

```python
# Can work with any data type
def my_name(name):
    print(name)

my_name('Kath')
```

#### Arbitrary Positional Argument

```python
#. The arguments are treated as tuples.
def number_args(*number):
    print(number[0] * number[1])

number_args(5,6,1,2,8)
```

```python
# Using tuple as argument
args_tuple = (5,6,1,2,8)

number_args(*args_tuple)
```

#### Keyword Argument

```python
def num_squared_cust(number, power):
    print(number ** power)
    
num_squared_cust(power = 5, number = 3)

```

#### Arbitrary Keyword Argument

```python
# The output of arbitraty keyword argument is a dictionary
def num_kwarg(**number):
    print(number)

num_kwarg(integer = '2309', integer2 = '349')
```

```python
def num_kwarg(**number):
    print('My first number is: ' +  number['integer'] + '\nMy other number is: ' + number['integer2'])

num_kwarg(integer = '2309', integer2 = '349')
```
